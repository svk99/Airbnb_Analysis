```python
from pymongo import MongoClient
import pandas as pd
```

```python
# Create a MongoClient
# client = MongoClient(connection_string)
cluster = MongoClient("mongodb+srv://stranger84:stranger84@cluster0.htcm4tk.mongodb.net/?retryWrites=true&w=majority")
print(cluster.list_database_names())
```

```python
# Get a reference to the database
db = cluster['sample_airbnb']
# Get a reference to a collection within the database
collection = db["listingsAndReviews"]
doc_list=[]
#loop through and add each document to list
for doc in collection.find():
  doc_list.append(doc)
df=pd.DataFrame(doc_list)
```

```python
#Find number of null values for each column
df.isnull().sum()
```

```python
#Filling null values
mode_list=['first_review','last_review','bedrooms','beds','bathrooms']
for i in mode_list:
  mode=df[i].mode()[0]
  df[i].fillna(mode,inplace=True)
```

```python
#Filling null values
mean_list=['security_deposit','cleaning_fee','weekly_price','monthly_price']
for i in mean_list:
  df[i]=df[i].apply(lambda x:float(str(x)))
  df[i].fillna(df[i].mean(),inplace=True)
```

```python
df['reviews_per_month'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
#Get coordinates and country from address column and add it in new column
df['coordinates'] = df['address'].apply(lambda x: x['location']['coordinates'])
df['country']=df['address'].apply(lambda x:x['country'])
```

```python
#Save the dataframe to csv file
df.to_csv('airbnb.csv')
```
